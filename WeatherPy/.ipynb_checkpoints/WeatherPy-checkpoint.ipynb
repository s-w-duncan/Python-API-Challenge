## Python API Challenge - Part I

#### Original Analysis was completed on April 5, 2020. The code is written so that each time the kernel is restarted and ran, updated weather information and a random set of cities will be produced. 

```python
# Import Dependancies
from config import key
import numpy as np
import pandas as pd
import requests
import time
from scipy.stats import linregress
from matplotlib import pyplot as plt
import pprint
import csv
import json
```

```python
# Load in city data
city_df = pd.read_csv("Resources/worldcities.csv")
city_random = city_df.sample(550)
cities = city_random["City"].tolist()
```

```python
# Request Weather Info
url = "http://api.openweathermap.org/data/2.5/weather?q="

# Create Lists to Store Data
town = []
lat = []
lon = []
temp = []
hum = []
cloud = []
wind = []

for city in cities:
    print(f"Proccessing Record {city}")
    
    response = requests.get(url + city + "&units=imperial&appid=" + key).json()
    
    try:
        town.append(response['name'])
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        hum.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        
    except KeyError:
        print("Cannot find key")
```

```python
# Compile and Create DataFrame
weather_dict = {
    "City": town,
    "Latitude": lat,
    "Longitude": lon,
    "Temperature (F)": temp,
    "Humidity (%)": hum,
    "Cloudiness (%)": cloud,
    "Wind Speed (mph)": wind}
weather_data = pd.DataFrame(weather_dict)
weather_data = weather_data.drop_duplicates(subset='City', keep="first")

# Save File to CSV
weather_data.to_csv("Outputs/Weather_Data.csv", index=False, header=True)

# Show DataFrame
weather_data
```

```python
# Scatter Plot Temperature (F) vs. Latitude
    
# Define Variables
x0 = weather_data["Temperature (F)"]
y = weather_data["Latitude"]
N = len(weather_data["City"].value_counts())
colors = np.random.rand(N)

# Create Plot and Labels
plt.scatter(x0, y, c=colors, alpha=0.5)
plt.xlabel("Temperature (F)")
plt.ylabel("Latitude")
plt.title("Latitude vs. Temperature (F)")

# Save Plot as PNG
plt.savefig("Outputs/Temp_Lat.png")
```

#### From the compiled weather data, each point in the scatter plot shows the temperature at the city's latitude. As latitude approaches 0, temperatures tend to increase.

```python
# Seperating City Data between Northern Hemisphere and Southern Hemisphere
north_lat = weather_data[y >= 0]
south_lat = weather_data[y < 0]
```

```python
# Northern Hemisphere (With Linear Regression) - Temperature (F) vs Latitude

# Define Variables
l_north = north_lat["Latitude"]
t_north = north_lat["Temperature (F)"]
N = len(north_lat["City"].value_counts())
colors = np.random.rand(N)

# Optional: Write Function For Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(t_north, l_north)
regress_values = t_north * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Create Plot and Labels
plt.scatter(t_north, l_north, c=colors, alpha=0.5)
plt.xlabel("Temperature (F)")
plt.ylabel("Latitude")
plt.title("Latitude vs. Temperature (F)")

# Plot Linear Regression
plt.plot(t_north,regress_values,"r-")

# Save Plot as PNG
plt.savefig("Outputs/North_Temp_Lat.png")
```

```python
# Southern Hemisphere (With Linear Regression) - Temperature (F) vs Latitude

# Define Variables
l_south = south_lat["Latitude"]
t_south = south_lat["Temperature (F)"]
N = len(south_lat["City"].value_counts())
colors = np.random.rand(N)

# Optional: Write Function For Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(t_south, l_south)
regress_values = t_south * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Create Plot and Labels
plt.scatter(t_south, l_south, c=colors, alpha=0.5)
plt.xlabel("Temperature (F)")
plt.ylabel("Latitude")
plt.title("Latitude vs. Temperature (F)")

# Plot Linear Regression
plt.plot(t_south,regress_values,"r-")

# Save Plot as PNG
plt.savefig("Outputs/South_Temp_Lat.png")
```

#### The linear regressions for the northern and southern hemispheres point towards a conclusion that as latitude approaches 0, on average, temperatures tend to increase. Outliers may be influenced by a number of different variables such as weather storms, or a city's geographical terrain.
